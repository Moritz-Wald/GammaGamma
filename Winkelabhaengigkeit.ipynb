```python
import matplotlib.pyplot as plt
import numpy as np  
```

```python
def W_theta_0_1_0(theta: np.ndarray) -> np.ndarray:
    return 1+np.cos(theta)**2
```

```python
def W_theta_4_2_0(theta: np.ndarray) -> np.ndarray:
    tmp = (16/207)* np.cos(2.*theta) + (1/207) * np.cos(4.*theta)
    return (115/112)*(1+tmp)
```

```python
theta = np.linspace(0, 2*np.pi, 1000, dtype=np.float64)
w1 = W_theta_0_1_0(theta)
w1 /= np.max(w1)
w2 = W_theta_4_2_0(theta)
w2 /= np.max(w2)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, subplot_kw={"projection": "polar"})
ax1.plot(theta, w1)
ax1.set_rmax(1)
ax1.set_title('0-1-0 Kaskade')
ax2.plot(theta, w2)
ax2.set_rmax(1)
ax2.set_title('4-2-0 Kaskade')
fig.suptitle('Winkelkorrelationsfunktionen')
plt.show()
fig.savefig('Winkelkorrelationsfunktionen.png')
```
